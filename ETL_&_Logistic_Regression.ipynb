```python
## extract data
## Enrollies
gg_sheetId = '1VCkHwBjJGRJ21asd9pxW4_0z2PWuKhbLR3gUHm-p4GI'
url = 'https://docs.google.com/spreadsheets/d/' + gg_sheetId + '/export?format=xlsx'
Enrollies_data = pd.read_excel(url, sheet_name='enrollies')
```
